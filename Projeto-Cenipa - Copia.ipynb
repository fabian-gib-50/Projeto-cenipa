### Criação de uma função para sumarizar variáveis importantes para anáse detalhada


```python
# Função para mapeamento das variáveis para procedimentos

def summary_missing_values(df):
    missing_values = pd.DataFrame(ocorrencia.isna().sum()/ocorrencia.shape[0])
    missing_values.reset_index(drop=True, inplace=True)    
    missing_values.columns = ["Absoluto dados ausentes(%)"]
    missing_values["Column"] = pd.Series(ocorrencia.columns).values
    # Make index be the column "Absolute Missing (%)"
    missing_values.reset_index(drop=True, inplace=True)
    missing_values.set_index("Column", inplace=True)
    
    # Calculando os valores comparando com o tatal de dados por coluna e os dados ausentes quantitativos em percentual
    column_missing = []
    column_total   = []
    column_miss    = []
    column_type    = []
    
    for column in ocorrencia.columns:
        col_miss = ocorrencia[column].isnull().sum()
        total_values = ocorrencia[column].count()       
        column_miss.append(col_miss)
        column_total.append(total_values) 
        column_missing.append((col_miss/total_values))
        column_type.append(column_type)
        
    # Construindo o dataframe que receberá os valores
    missing_values["Perc. Abs. ausentes(%)"] = pd.Series(column_missing).values
    missing_values["Perc. da equivalência(%)"] = -(pd.Series(column_missing).values-1)
    missing_values["Valores únicos"] = pd.Series(column_total).values
    missing_values["Valores absolutos"] = pd.Series(column_miss).values
    missing_values['Variáveis tipos'] = pd.DataFrame(ocorrencia.dtypes, columns=['column_type'])
    missing_values.sort_values(by="Absoluto dados ausentes(%)", ascending=False, inplace=True)
    missing_columns = list(missing_values.query("`Absoluto dados ausentes(%)` > 0").index)
    
    pd.options.display.float_format = '{:.2%}'.format
    return missing_values, missing_columns

missing_values, missing_columns_names = summary_missing_values(ocorrencia)
missing_values
```

```python
# Criando uma função para tratamento dos atributos "datetime" por faixa de horário das ocorrências
# Estruturando os binários de hora, minutos e segundos para criar um novo formato datetime
'''
  Função recebe a pd.Series com strings referentes a horários  e retorna 
  uma lista de períodos com os valores MANHÃ, TARDE, NOITE ou MADRUGADA
'''
def faixa_periodo_dia(series):
    periodos = []

    for valor in series:
      if not isinstance(valor, str):
         valor = '00:00:00'

      hora_valor     = int(valor[:2])
      minutos_valor  = int(valor[3:5])
      segundos_valor = int(valor[6:])
      hora = datetime.time(hora_valor,minutos_valor,segundos_valor)
    
      if datetime.time(0,0,0)    <= hora < datetime.time(5,59,59):
        periodos.append('MADRUGADA')
      elif datetime.time(6,0,0)  <= hora < datetime.time(11,59,59):
        periodos.append('MANHA')
      elif datetime.time(12,0,0) <= hora < datetime.time(17,59,59):
        periodos.append('TARDE')
      elif datetime.time(18,0,0) <= hora < datetime.time(23,59,59):
        periodos.append('NOITE')

    return periodos
```

```python
# Recebe a lista de períodos identificados a partir da lista de ocorencias 

lista_periodos = faixa_periodo_dia(ocorrencia['ocorrencia_hora'])

# Adiciona o atributo ocorrencia_periodo
ocorrencia['ocorrencia_periodos'] = pd.Series(lista_periodos)

# Removendo o atributo ocorrencia_hora
ocorrencia.drop(columns=['ocorrencia_hora'], inplace = True)
```

### Criação de uma função para plotagens dos gráficos e cenarizações.
  - Para cada plot, gráfico criado, chamo essa function def para tratar dos detalhamentos dos plots
   - Parâmetros plt %matplotlib para as labels 
   - Bar: gráficos verticais
   - Barh: gráficos horizontais
___

```python
### Function 
'''
  Dbian Data Scientist 
  Função responsável por anotar os indicadores das barras no gráfico.
  Entrada: tipo do gráfico ('bar' ou 'barh')
'''
def autolabel(tipo):
  if tipo == 'bar':
    for bar in plot.patches:
      plot.annotate(format(bar.get_height()),
                xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                xytext=(0, 8),
                textcoords='offset points',
                ha='center', 
                va='center',
                size=11)
  elif tipo == 'barh':
    for bar in plot.patches:
      plot.annotate(format(bar.get_width()),
                  xy = (bar.get_width() + 0.3, bar.get_y() + 0.38),
                  xytext=(15, -2),
                  textcoords='offset points',
                  ha='center', 
                  va='center',
                  size=11)
```
